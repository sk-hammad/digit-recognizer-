```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
dataset = pd.read_csv("/content/drive/MyDrive/personal/digit recognize /digit-recognizer/train.csv")
x = dataset.iloc[150:,1:].values
y = dataset.iloc[150:, 0].values
```

```python
validate_digit = dataset.iloc[:150, 1:].values
validate_label = dataset.iloc[:150, 0].values
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=0)
```

```python
x_train
```

```python
y_train
```

```python
x_train = x_train/255
x_test = x_test/255
validate_digit = validate_digit/255
```

```python
x_train = x_train.reshape(-1,28,28,1)
for i in range(5):
  plt.imshow(x_train[i])
  plt.title(y_train[i])
  plt.show()
```

```python
x_test = x_test.reshape(-1,28, 28,1)
for i in range(5):
  plt.imshow(x_test[i])
  plt.title(y_test[i])
  plt.show()
```

```python
from sklearn.preprocessing import LabelBinarizer
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten, Dense
```

```python
label_binarizer = LabelBinarizer()
y_train_one_hot = label_binarizer.fit_transform(y_train)
y_test_one_hot = label_binarizer.transform(y_test)
```

```python
model = Sequential()

model.add(Conv2D(32, (3, 3), activation='relu', input_shape=(28, 28, 1)))
model.add(MaxPooling2D((2, 2)))

model.add(Conv2D(64, (3, 3), activation='relu'))
model.add(MaxPooling2D((2, 2)))

model.add(Conv2D(64, (3, 3), activation='relu'))
model.add(Flatten())

model.add(Dense(64, activation='relu'))
model.add(Dense(10, activation='softmax'))
```

```python
model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])
history = model.fit(x_train, y_train_one_hot, epochs=7, validation_data=(x_test, y_test_one_hot))
```

```python
plt.plot(history.history['accuracy'], label='Training Accuracy')
plt.plot(history.history['val_accuracy'], label='Validation Accuracy')
plt.xlabel('Epochs')
plt.ylabel('Accuracy')
plt.legend()
plt.title('Training and Validation Accuracy')
plt.show()
```

```python
plt.plot(history.history['loss'], label='Training Loss')
plt.plot(history.history['val_loss'], label='Validation Loss')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.legend()
plt.title('Training and Validation Loss')
plt.show()
```

```python
model.save('digit.h5', model)
```

```python
from sklearn.metrics import accuracy_score
```

```python
validate_digit = validate_digit.reshape(-1, 28,28,1)
for i in range(5):
  plt.imshow(validate_digit[i])
  plt.title(validate_label[i])
  plt.show()
```

```python
predictions = model.predict(validate_digit)
predicted_labels = np.argmax(predictions, axis=1)
accuracy = accuracy_score(validate_label, predicted_labels)
print("Accuracy on test set:", accuracy)
```

```python
from sklearn.metrics import confusion_matrix, classification_report, f1_score
import seaborn as sns
plt.figure(figsize = (15, 8))
matr = confusion_matrix(validate_label, predicted_labels)
sns.heatmap(matr, annot = True, cmap = 'viridis')
plt.title('confusion matrix')
```

```python
print(classification_report(validate_label, predicted_labels))
```
